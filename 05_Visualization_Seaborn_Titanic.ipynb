# Seaborn Exercises



## The Data
visualization of the Titanic data with seaborn:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import scipy.stats as stats
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
jp = sns.jointplot(x="fare",y="age",data = titanic,kind="scatter")
jp.annotate(stats.pearsonr)
plt.show()
```

Interpretation: There is a weak negative correlation between age and fare. Makes some sense because older folks may have more financial power.

```python
sns.distplot(titanic["fare"],kde=False,bins=35)
```

```python
sns.boxplot(x="class",y="age",data=titanic)
```

Interpretation: Median age is highest for First class and lowest for third class, as we already saw in the joint plot.However, outliers in 2nd and 3rd class indicate that there were older folks in these classes and also that more older folks in 3rd class than 2nd class. Probably the survival rate of such older folks was very low (physically weak as well as lower lower survival rate of the 2nd/3rd class itself.)

```python
sns.swarmplot(x="class",y="age",data=titanic)
```

```python
sns.countplot(x="sex",data=titanic)
```

```python
corr = titanic.corr()
sns.heatmap(corr,vmin=-1.0,vmax=1.0)
```

```python
g = sns.FacetGrid(col="sex",data=titanic)
g.map(plt.hist,"age")
```
